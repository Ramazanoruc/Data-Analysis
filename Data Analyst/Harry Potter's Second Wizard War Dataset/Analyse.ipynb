```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = '/kaggle/input/harry-potters-second-wizard-war-dataset/Second Wizard War DeadorAliveUpdated.csv'
df = pd.read_csv(file_path)


df.head()
```

```python

gender_distribution = df['GENDER'].value_counts()


blood_status_distribution = df['BLOOD STATUS'].value_counts()


species_distribution = df['Species'].value_counts()


plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.barplot(x=gender_distribution.index, y=gender_distribution.values)
plt.title('Gender Distribution')

plt.subplot(1, 3, 2)
sns.barplot(x=blood_status_distribution.index, y=blood_status_distribution.values)
plt.title('Blood Status Distribution')

plt.subplot(1, 3, 3)
sns.barplot(x=species_distribution.index, y=species_distribution.values)
plt.title('Magical Species Distribution')

plt.show()
```

```python

house_count = df['HOUSE'].value_counts()


plt.figure(figsize=(8, 5))
sns.barplot(x=house_count.index, y=house_count.values)
plt.title('Distribution of Individuals in Hogwarts Houses')
plt.xlabel('Hogwarts House')
plt.ylabel('Count')
plt.show()
```

```python

survival_rate = df['ALIVE?'].value_counts(normalize=True) * 100


plt.figure(figsize=(6, 4))
sns.barplot(x=survival_rate.index, y=survival_rate.values)
plt.title('Overall Survival Rate')
plt.xlabel('Survival Status')
plt.ylabel('Percentage')
plt.show()


plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
sns.countplot(x='ALIVE?', hue='BLOOD STATUS', data=df)
plt.title('Survival by Blood Status')

plt.subplot(1, 2, 2)
sns.countplot(x='ALIVE?', hue='HOUSE', data=df)
plt.title('Survival by Hogwarts House')

plt.show()
```

```python

hair_color_distribution = df['HAIR COLOR'].value_counts()


common_boggarts = df['BOGGART'].value_counts().head(5)


patronus_distribution = df['PATRONUS'].value_counts()


plt.figure(figsize=(15, 5))

plt.subplot(1, 1, 1)
sns.barplot(x=hair_color_distribution.index, y=hair_color_distribution.values)
plt.title('Distribution of Hair Colors')
plt.show()

plt.subplot(2, 1, 1)
common_boggarts.plot(kind='bar')
plt.title('Most Common Boggart Forms')
plt.show()

plt.subplot(3, 1, 1)
sns.barplot(x=patronus_distribution.index, y=patronus_distribution.values)
plt.title('Patronus Distribution')

plt.show()
```

```python

nationality_distribution = df['Nationality'].value_counts()


plt.figure(figsize=(10, 6))
sns.barplot(x=nationality_distribution.index, y=nationality_distribution.values)
plt.title('Distribution of Nationalities')
plt.xlabel('Nationality')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


categorical_columns = ['HOUSE', 'BOGGART', 'PATRONUS'] 


df['HOUSE'] = df['HOUSE'].fillna('Unknown')


df['ALIVE?'] = df['ALIVE?'].fillna('Unknown')


plt.figure(figsize=(15, 5))
for i, col in enumerate(categorical_columns, 1):
    plt.subplot(1, len(categorical_columns), i)
    sns.countplot(x=col, hue='ALIVE?', data=df)
    plt.title(f'Count Plot of {col} by ALIVE?')
    plt.xticks(rotation=45, ha='right')

plt.show()
```

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
us_ev_data = pd.read_csv("Electric_Vehicle_Population_Data.csv")
```

1. What are the top 10 states with the largest EV populations?

```python
print(us_ev_data['State'].value_counts().head(10))
```

This data mostly collected in WA, so let see top ten city in WA


```python
WA_ev_popu = us_ev_data[us_ev_data['State'] == "WA"]
```

```python
state_counts = WA_ev_popu['City'].value_counts().head(10)
plt.figure(figsize=(10, 6))
plt.bar(state_counts.index, state_counts.values)
plt.xlabel('City')
plt.ylabel('Number of EVs')
plt.title('Top 10 Cities in WA with the largest EV populations')
plt.show()
```

2. Counties with the highest EV adoption rates

```python
county_counts = us_ev_data['County'].value_counts(normalize=True).head(10)
plt.figure(figsize=(10, 6))
plt.bar(county_counts.index, county_counts.values)
plt.xlabel('County')
plt.ylabel('EV adoption rate')
plt.title('Counties with the highest EV adoption rates')
plt.show()
```

3. EV makes most popular in the US

```python
make_counts = us_ev_data['Make'].value_counts().head(10)
plt.figure(figsize=(10, 6))
plt.bar(make_counts.index, make_counts.values)
plt.xlabel('Make')
plt.ylabel('Number of EVs')
plt.title('EV makes most popular in the US')
plt.show()
```

4. Trends in EV adoption over time

```python
model_year_counts = us_ev_data.groupby('Model Year')['VIN (1-10)'].count()
plt.figure(figsize=(10, 6))
plt.plot(model_year_counts.index, model_year_counts.values)
plt.xlabel('Model year')
plt.ylabel('Number of EVs')
plt.title('Trends in EV adoption over time')
plt.show()
```

5. EV models most popular in the US

```python
model_counts = us_ev_data['Model'].value_counts().head(10)
plt.figure(figsize=(10, 6))
plt.bar(model_counts.index, model_counts.values)
plt.xlabel('Model')
plt.ylabel('Number of EVs')
plt.title('EV models most popular in the US')
plt.show()
```
